```python
import re
import requests
import uuid
from PIL import Image
from io import BytesIO
import time
import cv2 as cv
import numpy as np
## this scripts only work for baidu image search

maxnumber = 20
## example find '钥匙'

urlhead = 'http://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word=手机&pn='
urltail = '&gsm=3c&ct=&ic=0&lm=-1&width=0&height=0'
teaturl = 'http://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word=%E9%92%A5%E5%8C%99&pn=0&gsm=3c&ct=&ic=0&lm=-1&width=0&height=0'


savepath = '~/Workspaces/tensorflow/tensorflow_learning/LostAndFound/data/card/'
```

```python
pattern = re.compile('objURL":"+?.*?"')
for i in range(maxnumber):
    url = urlhead + str(20 * i) + urltail
    r = requests.get(url)
    result = pattern.findall(r.content)
    time.sleep(0.25)
    print str(i) + 'in 600'
    for j in range(10):
        imgurl = result[j][9:]
        imgurl = imgurl[:-1]
        rimg = requests.get(imgurl)
        image = np.asarray(bytearray(rimg.content), dtype="uint8")
        img = cv.imdecode(image,1)
        cv.imwrite('other/' + str(uuid.uuid1()) + '.jpg', img)
```
